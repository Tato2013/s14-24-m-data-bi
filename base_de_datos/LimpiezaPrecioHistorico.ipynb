# Limpieza de precios que fueron mal ingresados y se agrega el precio en dolares para su uso

```python
import pandas as pd
```

```python
df=pd.read_csv('precios-historicos-argentina-2001-2024.csv')
```

```python
df
```

Trabajamos con datos de fecha para poder ordenarlos


```python
df['fecha_vigencia'] = pd.to_datetime(df['fecha_vigencia'],format='%d/%m/%Y')
```

```python
df.rename(columns={'fecha_vigencia': 'Fecha'}, inplace=True)
```

```python
df['precio'].max()
```

```python
df['precio'] = df['precio'].round(2)
```

```python
df = df.sort_values(by='Fecha').reset_index(drop=True)
```

```python
df
```

Se ven los ultimos valores de los precios para encontrar el valor mas alto

```python
df.tail(30)
```

Los valores mas altos estan entre 1300 y 1400

```python
filtar_precios=df[df['precio']>1400]
```

se crea un data frame con los valores mal ingresados para poder identificarlos

```python
filtar_precios
```

Para solucionarlo vamos a buscar precios del mismo dia y prodcuto para remplazarlos no es lo mas preciso pero da menos error y remplazarlo a mano llevaria mucho 
tiempo y es mejor que eliminar los datos

```python
for index, row in filtar_precios.iterrows():
    fecha = row['Fecha']
    id_producto = row['id_producto']
    precio_atipico = row['precio']
    
    # Filtrar el DataFrame original para encontrar valores similares basándose en la fecha e ID de producto
    valores_similares = df[(df['Fecha'] == fecha) & (df['id_producto'] == id_producto)]
    
    # Excluir los valores atípicos para calcular el máximo
    valores_similares_sin_atipicos = valores_similares[valores_similares['precio'] != precio_atipico]
    
    # Obtener el máximo valor de precio similar
    precio_maximo = valores_similares_sin_atipicos['precio'].max()
    
    # Reemplazar el valor atípico con el máximo valor encontrado
    df.loc[(df['Fecha'] == fecha) & (df['id_producto'] == id_producto) & (df['precio'] == precio_atipico), 'precio'] = precio_maximo
```

Se comprueba si se cambiaron los datos

# Lo que sigue es agregar los precios en dolares blue y oficial para realizar la comparacion

```python
dolarB=pd.read_csv('dolar_blue_normalizado.csv')
```

```python
dolarB.drop(columns='Unnamed: 0',inplace=True)
```

```python
dolarB['Fecha'] = pd.to_datetime(dolarB['Fecha'])
```

```python
dolarB.info()
```

```python
df.info()
```

Unimos los data frame por medio de la fecha para tomar la cotizacion y calcular los precios

```python
dolarB['Fecha'] = pd.to_datetime(dolarB['Fecha'])
dolarB = dolarB.set_index('Fecha')
dolarB = dolarB.resample('D').ffill().reset_index()

# Fusionar los DataFrames nuevamente
merged_df = pd.merge(df, dolarB, on='Fecha', how='left')

# Rellena los valores faltantes hacia adelante utilizando el último valor observado (método 'ffill')
merged_df['Venta'].fillna(method='ffill', inplace=True)

# Calcular el precio en dólares
merged_df['Precio_dolarBlue'] = merged_df['precio'] / merged_df['Venta']
```

```python
merged_df
```

Se buscara el valor del dolar para los datos faltantes y se remplazaran

```python
dolarB = dolarB.sort_values(by='Fecha').reset_index(drop=True)
```

```python
dolarB
```

```python
# Definir el valor del dólar en el año 2001
valor_dolar_2001 = 2.95

# Reemplazar los valores faltantes en la columna 'Venta' con el valor del dólar del año 2001
merged_df['Venta'].fillna(valor_dolar_2001, inplace=True)

# Calcular el precio en dólares
merged_df['Precio_dolarBlue'] = merged_df['precio'] / merged_df['Venta']
```

Renombramos columnas  para su posterior trabajo 

```python
merged_df.rename(columns={'Venta': 'Cotizacion_DolarBlue'}, inplace=True)
merged_df.rename(columns={'precio': 'Precio Ars'}, inplace=True)
```
