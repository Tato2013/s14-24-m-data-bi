## Corrección de los datos

Se detectaron errores en los valores desde el 2001 a 2016, por lo cual se buscaron los valores correctos para cada periodo.

Ahora se prodera a hacer el cambio y correccion manual de cada uno

```python
precios_actualizados = [0.54, 1.09, 0.60, 1.18, 1.19, 1.25, 1.58, 1.28, 1.55, 1.42, 1.38, 1.42, 1.57, 1.27, 1.18, 1.63, 1.35, 1.48, 1.68, 1.35, 2.20, 2.00, 2.10, 2.50, 2.40, 2.10, 2.00, 2.40, 2.10, 2.50, 2.00, 2.40, 0.75, 1.95, 1.60, 1.68, 1.55, 2.20, 2.60, 3.00, 3.10, 2.70, 3.10, 3.50, 3.50, 3.70, 3.80, 4.10, 5.00, 4.50, 5.50, 5.40, 5.10, 4.70, 4.80, 5.00, 5.40, 4.80, 4.70, 4.70, 5.40, 5.10, 4.80, 5.10, 4.80, 4.70, 5.40, 12.00, 6.00, 12.00, 5.30, 5.10, 15.69]
```

```python
for i in range(len(precios_actualizados)):
    df.loc[i, 'Precio Ars'] = float(precios_actualizados[i])
    df.loc[i, 'Precio_dolarBlue'] = float(df.loc[i, 'Precio Ars'] / df.loc[i, 'Cotizacion_DolarBlue'])
    df.loc[i, 'Precio_dolarOficial'] = float(df.loc[i, 'Precio Ars'] / df.loc[i, 'Cotizacion_DolarOficial'])

df
```

```python
df['Precio Ars'] = round(df['Precio Ars'], 2)
df['Precio_dolarBlue'] = round(df['Precio_dolarBlue'], 2)
df['Precio_dolarOficial'] = round(df['Precio_dolarOficial'], 2)

df
```

```python
precios_promedio = df.groupby(["Fecha", "id_producto"])["Precio Ars"].mean()

for index, row in df.iterrows():
    if row["Precio Ars"] == 0:
        fecha = row["Fecha"]
        id_producto = row["id_producto"]
        if (fecha, id_producto) in precios_promedio:
            df.at[index, "Precio Ars"] = precios_promedio[(fecha, id_producto)]
```

```python
df["Precio_dolarBlue"] = df["Precio Ars"] / df["Cotizacion_DolarBlue"]
df["Precio_dolarOficial"] = df["Precio Ars"] / df["Cotizacion_DolarOficial"]
```

```python
df.to_parquet('./archivos_csv/Historico precios con dolar.parquet', index=False)
```
